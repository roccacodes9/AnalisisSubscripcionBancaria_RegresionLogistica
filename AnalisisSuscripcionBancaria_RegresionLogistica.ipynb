## Import the relevant libraries

```python
# Importamos las bibliotecas necesarias
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
from scipy import stats

# Configuramos Seaborn para tener gráficos estéticamente agradables
sns.set()
```

## Load the data

Load the ‘Bank_data.csv’ dataset.

```python
# Cargamos los datos del archivo CSV
raw_data = pd.read_csv('Bank-data.csv')
raw_data
```

```python
# Creamos una copia de los datos originales para evitar alteraciones accidentales
data = raw_data.copy()

# Eliminamos la columna de índice que viene con los datos
data = data.drop(['Unnamed: 0'], axis=1)

# Mapeamos los valores 'yes' a 1 y 'no' a 0 en la columna 'y'
data['y'] = data['y'].map({'yes': 1, 'no': 0})
data
```

```python
data.describe()
```

### Declare the dependent and independent variables

Use 'duration' as the independet variable.

```python
# Seleccionamos las variables independientes y la variable dependiente
y = data['y']
x1 = data['duration']
```

### Simple Logistic Regression

Run the regression.

```python
# Agregamos una constante a las variables independientes para el modelo de regresión
x = sm.add_constant(x1)
# Creamos un modelo de regresión logística utilizando el método de Máxima Verosimilitud
reg_log = sm.Logit(y, x)
# Ajustamos el modelo a los datos
results_log = reg_log.fit()
```

### Interpretation

```python
# Mostramos un resumen de los resultados del modelo
results_log.summary()
```

The dependent variable is 'duration'. The model used is a Logit regression (logistic in common lingo), while the method - Maximum Likelihood Estimation (MLE). It has clearly converged after classifyin 518 observations.

The Pseudo R-squared is 0.21 which is within the 'acceptable region'.

The duration variable is significant and its coefficient is 0.0051.

The constant is also significant and equals: -1.70
